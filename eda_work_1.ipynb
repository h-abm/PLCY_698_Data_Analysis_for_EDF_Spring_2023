```python
import matplotlib.pyplot as plt
from datascience import *
import numpy as np
import pandas as pd
import matplotlib
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')

# load the data
final_data = pd.read_csv('EDF_Database_2v3_1.csv')
final_data.head(5)

final_data.dropna(inplace=True)
```

```python
# Check if there are any missing values
final_data.isnull().sum()
```

```python
# histogram of the total population bins of place fips
plt.hist(final_data['Total Population'], bins="auto")
plt.show()
```

```python
# correlation of total population and percent of land in 500-year floodplain
five_h_cor = final_data['Total Population'].corr(final_data['Percent of Land in 500-Year Floodplain'])
five_h_cor
```

```python
one_h_cor = final_data['Total Population'].corr(final_data['Percent of Land in 100-Year Floodplain'])
one_h_cor
```

```python
poverty_cor = final_data['Living in Poverty.2'].corr(final_data['Social Vulnerability (CDC Index)'])
poverty_cor
```

```python
# scatter plot of total population and percent of land in 500-year floodplain
plt.scatter(final_data['Total Population'], final_data['Percent of Land in 500-Year Floodplain'])
plt.show()
```

```python
# Correlation of total population and Total Percent of Land at Some Level of Flood Risk
total_cor = final_data['Rural Capacity (Headwaters Index)'].corr(final_data['Total Percent of Land at Some Level of Flood Risk'])
total_cor
```
